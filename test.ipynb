```python
#gpt废稿

#递归搜索配方
import yaml

# 读取并解析YAML文件
with open('resources.yaml', 'r') as file:
    recipes_data = yaml.safe_load(file)

def find_recipe(item_id):
    for item, data in recipes_data['resources'].items():
        if item == item_id:
            return data.get('recipes', [])
    #print(recipes_data.get('requirement_groups').items())
    # 如果在resources中没有找到，尝试在requirement_groups中查找
    for group, items in recipes_data.get('requirement_groups').items():
        if item_id ==group:
            return [{'recipe_type': 'Raw Resource'}]  # 对于requirement_groups中的物品，直接返回"Raw Resource"

def get_raw_materials(item_id, quantity=1):
    recipe = find_recipe(item_id)
    #print(recipe)

    if recipe:
        for step in recipe:
            if step['recipe_type'] == 'Raw Resource':
                return {item_id: quantity}
            else:
                requirements = step.get('requirements', {})
                raw_materials = {}
                for material, amount in requirements.items():
                    material_materials = get_raw_materials(material, abs(amount))
                    for mat, mat_qty in material_materials.items():
                        if mat in raw_materials:
                            raw_materials[mat] += mat_qty * quantity
                        else:
                            raw_materials[mat] = mat_qty * quantity

                return raw_materials

    return None
    

# 示例用法
user_input = 'Acacia Slab'
quantity = 1
materials_needed = get_raw_materials(user_input,quantity)

if materials_needed:
    print(f"要制作{quantity}个{user_input}，你需要的原材料是：")
    for material, quantity in materials_needed.items():
        print(f"{material}: {quantity}")
else:
    print(f"未找到 {user_input} 的配方。")
print(materials_needed)
```
